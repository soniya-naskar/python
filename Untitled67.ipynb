```python
#1.Polynomial Functions and Kernel Functions are closely related in the context of machine learning. The polynomial kernel is a specific type of kernel function that implicitly maps input features into a higher-dimensional space using polynomial functions.
#This mapping enables algorithms like SVMs to find non-linear decision boundaries without explicitly transforming the data into that high-dimensional space. The degree of the polynomial controls the complexity of this transformation, providing a balance between model flexibility and computational feasibility

#2.To implement an SVM with a polynomial kernel using scikit-learn, you can use the SVC class from the sklearn.svm module and specify the kernel as 'poly'.

#3.In Support Vector Regression (SVR), the parameter ϵ (epsilon) plays a crucial role in defining the margin of tolerance within which no penalty is given to errors. This means that predictions within the ϵ distance from the actual values are considered acceptable and do not contribute to the loss function.

#4.The performance of Support Vector Regression (SVR) is highly influenced by the choice of kernel function,C parameter, ϵ parameter, and γ parameter. Each of these parameters plays a specific role in controlling the model’s flexibility, complexity, and ability to generalize.
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
df= sns.load_dataset('iris')
df
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['species']= label.fit_transform(df['species'])
```

```python
fig,ax= plt.subplots(figsize=(15,10))
sns.boxplot(data=df, width=0.5, ax=ax, fliersize=3)
```

```python
X=df.drop(columns=['species'])
y= df['species']
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features (important for SVMs)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
fig,ax= plt.subplots(figsize=(15,10))
sns.boxplot(data=X_train, width=0.5, ax=ax, fliersize=3)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
C = 1.0  # Regularization parameter
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
y_pred=svm_model.predict(X_test)
acc=accuracy_score(y_test,y_pred)
acc
```

```python
param_grid={
    'C':[0.1,0,1,10],
    'kernel':['linear','rbf','poly'],
    'gamma': [0.1,0, 1, 10]
}
```

```python
from sklearn.model_selection import GridSearchCV
gridsearch=GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
gridsearch
```

```python
gridsearch.fit(X_train, y_train)
```

```python
gridsearch.best_params_
```

```python
gridsearch.best_score_
```

```python
best_svc = gridsearch.best_estimator_
X_scaled = scaler.fit_transform(X)
best_svc.fit(X_scaled, y)
```

```python
y_pred=svm_model.predict(X_test)
acc=accuracy_score(y_test,y_pred)
acc
```
