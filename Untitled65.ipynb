```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
df= pd.read_csv("diabetes.csv")
df
```

```python
#1.
df.describe()
```

```python
df.info()
```

```python
#2.
df.isnull().sum()
```

```python
df['BMI']=df['BMI'].replace(0,df['BMI'].mean())
df['Glucose']=df['Glucose'].replace(0,df['Glucose'].mean())
df['BloodPressure']=df['BloodPressure'].replace(0,df['BloodPressure'].mean())
df['SkinThickness']=df['SkinThickness'].replace(0,df['SkinThickness'].mean())
df['Insulin']=df['Insulin'].replace(0,df['Insulin'].mean())
```

```python
df.describe()
```

```python
#3.
X=df.drop(columns= ['Outcome'])
y=df['Outcome']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=0)
X_train.shape, X_test.shape
```

```python
#4.
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score
from sklearn.model_selection import GridSearchCV
decision= DecisionTreeClassifier()
param_grid = {
    'criterion': ['gini', 'entropy'],  # 'entropy' for C4.5 like behavior
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
```

```python
grid_search = GridSearchCV(decision, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
best_dt_model = DecisionTreeClassifier(**grid_search.best_params_)
```

```python
best_dt_model
```

```python
best_param=grid_search.best_params_
best_dt_model.fit(X_train,y_train)
```

```python
#5.
# Predict using the test set
y_pred = best_dt_model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate precision
precision = precision_score(y_test, y_pred, average='weighted')

# Calculate recall
recall = recall_score(y_test, y_pred, average='weighted')

# Calculate F1 score
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
accuracy,precision,recall,f1
```

```python
best_dt_model.best_score_
```

```python
con = confusion_matrix(y_pred, y_test)
con
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(con, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
predictions = grid_search.predict_proba(X_test)
auc = roc_auc_score(y_test, predictions[:,1])

print (auc)

fpr, tpr, _ = roc_curve(y_test, predictions[:,1])

plt.clf()
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve')
plt.show()
```
